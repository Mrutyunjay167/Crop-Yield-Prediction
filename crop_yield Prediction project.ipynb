EXPLANATORY DATA ANALYSIS

```python
crop_p=df.groupby('Crop')[['Production']].sum()
crop_production=crop_p.sort_values(by=['Production'],ascending=False).head(10)
crop_production
```

```python
from plotly import plot
fig=px.bar(crop_production,x='Production',y=crop_production.index,log_x=True,title='Top 10 Crop <b>Produced</b> in India')
fig.update_yaxes(autorange="reversed")
fig.update_layout(xaxis_title='Production',yaxis_title='Crop')
fig.update_traces(marker_color='Peru',marker_line_color='slategray',marker_line_width=1,orientation='h')
fig.update_layout(title_font_color='sienna',plot_bgcolor='white')
fig.show()
```

The graph indicates the top 10 crops produced in India. Coconut, sugarcane and rice are the top three produced crops, where production of coconut surpasses other crops with a huge amount of 308.752B. While the production of sugarcane and rice are 7.236B and 2.229B followed by wheat production which accounts to 1.811B.

```python
coco_p = df[df["Crop"] == 'Coconut ']
coco_p.head()
```

```python
fig=px.scatter(data_frame=coco_p, x='Crop_Year',y='Production',color='State',title='Year of <b>Coconut Produced</b> in different State')
fig.update_xaxes(title='Year')
fig.update_traces(marker_line_width=1,marker_size=10)
fig.update_layout(title_font_color='Brown')
fig.show()
```

The graph indicates that Kerala is the highest coconut producing state followed by Karataka and Tamil Nadu from 1997 to 2019. Although there are some exceptions, in 2011 and 2015 Tamil Nadu surpassed Kerala in coconut production while in 2019 Karnataka was the highest coconut producing state followed by Tamil Nadu and Kerala. The graph also shows that in 1997 there was no production of coconut in Karnataka. In Tamil Nadu also there was no production of coconut before 2002, there was also no production in 2010 and 2012.

Production of coconut with slight increase in all the years is noticed in West Bengal and Assam while Kerala has increase and decrease in few years. Andhra Pradesh which is the fourth coconut producing state in India had no production before 2000, after which production increased in all the years except in 2007 when there was no production.

TRAIN_TEST_SPLIT

```python
X = df.drop(['Crop', 'Crop_Year', 'Season', 'State', 'Yield'], axis = 1)
y = df['Yield']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

MODEL BUILDING

LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
lr.score(x_test,y_test)
```

```python
lr.score(x_train,y_train)
```

```python
lr_predict=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
r2=r2_score(y_test,lr_predict)
r2
```

```python
fig=px.scatter(x=y_test,y=lr_predict,trendline='ols',title='<b>LINEAR REGRESSION</b>')
fig.update_traces(marker_size=10,marker_line_width=1,marker_color='sienna')
fig.update_xaxes(title='Yield')
fig.update_yaxes(title='Production')
fig.show()
```

RIDGE REGRESSION (GRID SEARCH CV)

```python
from sklearn.linear_model import Ridge
ridge=Ridge()
from sklearn.model_selection import GridSearchCV
param_grid={'alpha': [0.001,0.1,1,10,100]}
grid=GridSearchCV(ridge,param_grid,cv=5,scoring='r2')
grid.fit(x_train,y_train)
```

```python
grid.score(x_test,y_test)
```

```python
grid_predict=grid.predict(x_test)
```

```python
r2
```

```python
grid.best_estimator_.coef_
```

```python
grid.best_estimator_.intercept_
```

```python
grid.best_estimator_
```

```python
fig=px.scatter(x=y_test,y=grid_predict,trendline='ols',title='<b>RIDGE REGESSION (GridSearch CV)</b>')
fig.update_traces(marker_size=12,marker_line_width=1,marker_color='sienna')
fig.update_xaxes(title='Yield')
fig.update_yaxes(title='Prediction')
fig.show()
```
